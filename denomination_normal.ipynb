```python
import pygame
import sys

# Configuration
CELL_SIZE = 60
GRID_COLOR = (40, 40, 40)
BACKGROUND_COLOR = (255, 255, 255)
NULL_COLOR = (100, 100, 100)
TEXT_COLOR = (0, 0, 0)
COLORS = {
    1: (200, 200, 200),  # Color for null piece
    2: (255, 200, 200),  # Color for piece with denomination 2
    3: (200, 255, 200),  # Color for piece with denomination 3
    4: (200, 200, 255),  # Color for piece with denomination 4
    5: (255, 255, 200)   # Color for piece with denomination 5
    # Add more colors for higher denominations as needed
}

def get_grid_size():
    while True:
        try:
            n = int(input("Enter grid size (3-8 recommended): "))
            if 2 <= n <= 10:
                return n
            print("Please enter a number between 2 and 10")
        except ValueError:
            print("Please enter a valid integer")

class NumberGridGame:
    def __init__(self, n):
        self.n = n
        self.grid = [[0 for _ in range(n)] for _ in range(n)]
        self.placed = {1}  # Keep track of placed denominations
        self.current_max = 1  # Maximum denomination placed so far
        self.init_pygame()

    def init_pygame(self):
        pygame.init()
        self.screen = pygame.display.set_mode((self.n * CELL_SIZE, self.n * CELL_SIZE))
        pygame.display.set_caption("Denomination Grid Game")
        self.font = pygame.font.Font(None, 36)

    def adjacent_sum(self, row, col):
        total = 0
        for i in [-1, 0, 1]:
            for j in [-1, 0, 1]:
                if i == 0 and j == 0:
                    continue
                r, c = row + i, col + j
                if 0 <= r < self.n and 0 <= c < self.n:
                    total += self.grid[r][c]
        return total

    def can_place_null(self, row, col):
        # Check if all adjacent cells are empty for placing a null piece
        for i in [-1, 0, 1]:
            for j in [-1, 0, 1]:
                if i == 0 and j == 0:
                    continue
                r, c = row + i, col + j
                if 0 <= r < self.n and 0 <= c < self.n:
                    if self.grid[r][c] != 0:
                        return False
        return True

    def handle_click(self, row, col):
        if self.grid[row][col] != 0:
            return  # Cell already occupied

        adj_sum = self.adjacent_sum(row, col)

        # Case 1: Place null piece (denomination 1)
        if adj_sum == 0:
            if self.can_place_null(row, col):
                self.grid[row][col] = 1
                return True  # Placement successful

        # Case 2: Place higher denomination piece
        else:
            k = adj_sum  # Denomination to be placed
            if k > 1 and all(d in self.placed for d in range(1, k)):
                # Check if all lower denominations have been placed
                self.grid[row][col] = k
                self.placed.add(k)
                self.current_max = max(self.current_max, k)
                return True  # Placement successful
        return False  # Placement failed

    def draw_grid(self):
        self.screen.fill(BACKGROUND_COLOR)  # Clear the screen

        for row in range(self.n):
            for col in range(self.n):
                val = self.grid[row][col]  # Value of the cell
                rect = pygame.Rect(col * CELL_SIZE, row * CELL_SIZE, CELL_SIZE, CELL_SIZE)

                # Draw cell background based on value
                if val == 0:
                    color = BACKGROUND_COLOR
                else:
                    color = COLORS.get(val, BACKGROUND_COLOR)
                pygame.draw.rect(self.screen, color, rect)
                pygame.draw.rect(self.screen, GRID_COLOR, rect, 1)  # Draw grid lines

                # Draw value or null piece representation
                if val == 1:
                    center = (col * CELL_SIZE + CELL_SIZE // 2, row * CELL_SIZE + CELL_SIZE // 2)
                    pygame.draw.circle(self.screen, NULL_COLOR, center, CELL_SIZE // 4)  # Null piece as a circle
                elif val > 1:
                    text = self.font.render(str(val), True, TEXT_COLOR)
                    text_rect = text.get_rect(center=(col * CELL_SIZE + CELL_SIZE // 2, row * CELL_SIZE + CELL_SIZE // 2))
                    self.screen.blit(text, text_rect)  # Display denomination number

    def run(self):
        while True:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    pygame.quit()
                    sys.exit()

                if event.type == pygame.MOUSEBUTTONDOWN:
                    x, y = pygame.mouse.get_pos()
                    col = x // CELL_SIZE
                    row = y // CELL_SIZE
                    self.handle_click(row, col)  # Handle player's click

            self.draw_grid()  # Redraw the grid
            pygame.display.flip()  # Update the display

if __name__ == "__main__":
    n = get_grid_size()
    game = NumberGridGame(n)
    game.run()
```
